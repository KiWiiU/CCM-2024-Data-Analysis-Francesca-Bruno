## Question 2 (student-made): Which age groups were most responding to the top 5 most influential activities/events?

### Determining how many times each activity was described as having a 'high impact' for a non-major student

```python
influence = {}
for i in range(7, 18):
    influence[i] = df[df.columns[i]].value_counts()["High Impact"]
```

### Finding the top 5 most influential activites and sorting them in a list from most to least impactful

```python
influenceRanked = []
for x in range(5):
    highest = 0
    index = 0
    for y in range(7, 18):
        if influence[y] > highest:
            index = y
            highest = influence[y]
    influenceRanked.append(df.columns[index])
    influence[index] = 0
print(influenceRanked)
```

### Isolating rows in the data frame where a specific activity/influence was described as having a high impact, and taking the mode (most common) age value

```python
dfFF = df[df['familyFriendInfluence'] == "High Impact"]
print("The age group most influenced by family and friends is " + str(dfFF['age'].mode()))

dfEmployer = df[df['employer'] == "High Impact"]
print("The age group most influenced by employers is " + str(dfEmployer['age'].mode()))

dfClass = df[df['schoolCompsciClass'] == "High Impact"]
print("The age group most influenced by middle/high school computing classes is " + str(dfClass['age'].mode()))

dfFFWork = df[df['familyFriendWork'] == "High Impact"]
print("The age group(s) most influenced by family or friends who work in the computer science field is " + str(dfFFWork['age'].mode()))

dfTeacher = df[df['highschoolTeacher'] == "High Impact"]
print("The age group most influenced by a highschool teacher is " + str(dfTeacher['age'].mode()))
```

## Question 4 (student-made): What is the most popular class among highschoolers (18 or under students)?

### Determining how many students who are 18 years old or younger are taking each of the three computing classes listed on the survey by isolating the data from students 18 or younger and finding the amount of times each class appears in the new dataframe

```python
classPopularity = {}
dfKid = df[df['age'] == "18 and younger"]
classPopularity["CMP126"] = dfKid["course"].value_counts()["CMP 126 Computer Technology and Applications"]
classPopularity["CMP101"] = dfKid["course"].value_counts()["CMP 101 Computer Information Literacy"]
classPopularity["CMP135"] = dfKid["course"].value_counts()["CMP 135 Computer Concepts with Applications"]
if classPopularity["CMP101"] > classPopularity["CMP126"] and classPopularity["CMP135"] < classPopularity["CMP101"]:
    print("The most popular class for highschool students (students 18 or under) is CMP101 Computer and Information Literacy")
if classPopularity["CMP135"] > classPopularity["CMP126"] and classPopularity["CMP135"] > classPopularity["CMP101"]:
    print("The most popular class for highschool students (students 18 or under) is CMP135 Computer Concepts with Applications")
if classPopularity["CMP126"] > classPopularity["CMP135"] and classPopularity["CMP126"] > classPopularity["CMP101"]:
    print("The most popular class for highschool students (students 18 or under) is CMP126 Computer Concepts with Applications")
```

### The data analysis from these four questions is useful for increasing enrollment for computing classes at CCM because determining the most impactful influences, most popular classes, and most predominant age group can help with advertising the classes because it not only gives a target audience, but also predicts which classes and activities are most likely to be successful in increasing enrollment. Possible insights that can be drawn from this data are the fact that the vast majority of students taking computing classes at CCM are young adults, Computer Concepts with Applications is the most popular class among highschool students, and the most predominant age group in computing classes is most influenced by factors outside of CCM, such as family members, friends, and highschool teachers.
